```python
import numpy as np
import matplotlib.pyplot as plt
import glob
import pandas as pd

num_sim = 100

track_pool1 = np.zeros((num_sim,11))
track_pool2 = np.zeros((num_sim,11))
track_pool3 = np.zeros((num_sim,11))
track_pool4 = np.zeros((num_sim,11))

all_track_pool1 = glob.glob('.csv') #100% cov
all_track_pool2 = glob.glob('.csv') #75% cov
all_track_pool3 = glob.glob('.csv') #50% cov
all_track_pool4 = glob.glob('.csv') #25% cov


list_data_NP1 = []
#
for filename in all_track_pool1:
     data_NP1 = pd.read_csv(filename,header=None)
     list_data_NP1.append(data_NP1)

pd.concat(list_data_NP1,ignore_index=True)

for i in range(0,num_sim):
    track_pool1[[i],:] = list_data_NP1[i].iloc[0:11].T


mean_track_pool1 = np.mean(track_pool1, axis = 0)
std_track_pool1 = np.std(track_pool1, axis = 0)




#######

list_data_NP2 = []
#
for filename in all_track_pool2:
     data_NP2 = pd.read_csv(filename,header=None)
     list_data_NP2.append(data_NP2)

pd.concat(list_data_NP2,ignore_index=True)

for i in range(0,num_sim):
    track_pool2[[i],:] = list_data_NP2[i].iloc[0:11].T


mean_track_pool2 = np.mean(track_pool2, axis = 0)
std_track_pool2 = np.std(track_pool2, axis = 0)

#######

list_data_NP3 = []
#
for filename in all_track_pool3:
     data_NP3 = pd.read_csv(filename,header=None)
     list_data_NP3.append(data_NP3)

pd.concat(list_data_NP3,ignore_index=True)

for i in range(0,num_sim):
    track_pool3[[i],:] = list_data_NP3[i].iloc[0:11].T


mean_track_pool3 = np.mean(track_pool3, axis = 0)
std_track_pool3 = np.std(track_pool3, axis = 0)

#######

list_data_NP4 = []
#
for filename in all_track_pool4:
     data_NP4 = pd.read_csv(filename,header=None)
     list_data_NP4.append(data_NP4)

pd.concat(list_data_NP4,ignore_index=True)

for i in range(0,num_sim):
    track_pool4[[i],:] = list_data_NP4[i].iloc[0:11].T

mean_track_pool4 = np.mean(track_pool4, axis = 0)
std_track_pool4 = np.std(track_pool4, axis = 0)

cycles = 10
t_cycles = np.linspace(0,cycles,cycles+1)
# pool size
fig,ax = plt.subplots()
ax.plot(t_cycles,mean_track_pool1, marker = 'o',color = 'xkcd:black')
ax.plot(t_cycles,mean_track_pool2, marker = 'o',color = 'xkcd:dark grey')
ax.plot(t_cycles,mean_track_pool3, marker = 'o',color = 'xkcd:grey')
ax.plot(t_cycles,mean_track_pool4, marker = 'o',color = 'xkcd:silver')
ax.set_yscale('log')
plt.ylim((0.1,1000000+1000000))
plt.show()

time_ext = []
i_counter = []
i_noext_counter = []

for i in range(0,num_sim):
    for j in range(1,11):
        if track_pool4[i,j-1] != 0 and track_pool4[i,j] == 0:
            time_ext.append(j)
            i_counter.append(i)
    if track_pool4[i,-1] != 0:
        i_noext_counter.append(i)
        
print('time ext shape',np.shape(time_ext))
print('i counter shape',np.shape(i_counter))
print(' i noext shape', np.shape(i_noext_counter))
```
